```python
#B
def grahm_schmidt(A):
    m, n = A.shape
    Q = np.zeros((m, n))  
    R = np.zeros((n, n))  

    for j in range(n):
        v = A[:, j]  
        for i in range(j):
            R[i, j] = np.dot(Q[:, i], A[:, j])  
            v = v - R[i, j] * Q[:, i]  

        R[j, j] = np.linalg.norm(v)  
        Q[:, j] = v / R[j, j] 
    return Q, R

base=grahm_schmidt(A)
base
```

```python
import numpy as np

def g_s(A):
    m, n = A.shape
    Q = np.zeros((m, n))  
    R = np.zeros((n, n))

    for i in range(n):
        v = A[:, i]
        for j in range(i):
            R[j, i] = np.dot(Q[:, j], v)
            v = v - R[j, i] * Q[:, j]
        R[i, i] = np.linalg.norm(v)
        Q[:, i] = v / R[i, i]

    return Q, R

A = np.array([[3, 1, -1], [1, 2, 0], [0, 1, 2], [1, 1, -1]])
g_s(A)
```

```python
#markov 1

#c

import matplotlib.pyplot as plt
from itertools import permutations
import numpy as np
from itertools import combinations_with_replacement

Pj= np.array([[0.5, 0.5]]) # cara sello
Ps= np.array([[0.9, 0.1]])

T = np.array([[0.8, 0.2],
              [0.2,0.8]])
E = np.array([[0.5, 0.9],
              [0.5,0.1]])

States= np.array([0,1]) #0 S, 1 C

DictH = {0:'Sello',1:'Cara'} 

apri=np.array([0.2, 0.8]) #para ps y pj

Obs= np.array([0,1,1,1,0,1,0,1])
#(S,C,C,C,S,C,S,C)
```

```python
def GetStates(States,N):
    
    CStates = list( combinations_with_replacement(States,N) )
    
    print(CStates)
    
    Permu = []
    
    for it in CStates:
        p = list(permutations(it,N))
       # print(p)
        
        for i in p:
            if i not in Permu:
                Permu.append(i)
                
    return np.array(Permu)
```

```python
HiddenStates = GetStates(States,8)
HiddenStates
```

```python
def GetProb(T,E,Obs,State,Prior):
    
    n = len(Obs)
    p = 1.
    
    p *= Prior[State[0]]
    
    # Matriz de transicion
    for i in range(n-1):
        p *= T[ State[i+1], State[i]  ]
        
    for i in range(n):
        p *= E[ Obs[i], State[i] ]
        
    return p
```

```python
P = np.zeros(HiddenStates.shape[0], dtype=np.float64)
```

```python
for i in range(P.shape[0]):
    P[i] = GetProb(T,E,Obs,HiddenStates[i],apri)
    
print(P)
```

```python
plt.plot(P)
```

```python
ii = np.where( P == np.amax(P))
print(f"El estado oculto m√°s probable es: {HiddenStates[ii]}, es decir:")
lista = []
for i in HiddenStates[ii][0]:
    lista.append(DictH[i])
print(lista)
print(f"Y su probabilidad es de : {max(P)}")
```

```python
ObsStates = GetStates([0,1],8)
Nobs = ObsStates.shape[0]

PObs = np.zeros(Nobs)

for j in range(Nobs):
    
    dim = HiddenStates.shape[0]
    P = np.zeros(dim)
    
    for i in range(dim):
        P[i] = GetProb(T,E,ObsStates[j],HiddenStates[i],apri)
        
    PObs[j] = np.sum(P)
```

```python
ii = np.where( PObs == np.amax(PObs))
print(ObsStates[ii])
```
